```python
#imported libraries for data wrangling
import pandas as pd
import json
import ast
```

```python
cp = pd.read_csv("compounds.csv", index_col=0)
cp #the data for compounds, contains data for associations between compounds and the bioassays that test them
```

```python
bs = pd.read_csv("assays.csv", index_col=0)
bs = bs.drop(bs.columns[0], axis=1)
bs #the data for bioassays
```

```python
a2g = pd.read_csv("genes.csv", index_col = 0)
a2g #the data for bioassays, contain within the JSON the association from bioassay to the gene it interacts with
```

```python
#only interested in these two columns to find association
cp2 = cp[['cid','aids']]
cp2
```

```python
#want to make a new dataframe that makes a new row for every mapping of cid to aid, 'c2a'
c2a = pd.DataFrame(columns=['cid', 'aid'])
for num in range(0, len(cp2)):
    c, a = cp2.iloc[num]
    a = json.loads(a)
    if (a != []):
        for each_a in a:
            row_to_add = pd.Series({'cid': c, 'aid': each_a})
            c2a = c2a.append(row_to_add, ignore_index=True)
c2a            
```

```python
c2a.to_csv("compounds_assays_associations.csv", index=False)
```

```python
#this is the format of code that can properly parse the JSON object that an assay contains to find the relevant info
ast.literal_eval(a2g.iloc[7]['JSON'])['InformationList']['Information'][0]
```

```python
#want to make a new dataframe that makes a new row for every mapping of aid to gid, 'aid2gid'
aid2gid = pd.DataFrame(columns=['aid', 'GeneID'])
for num in range(1, len(a2g)):
    a, Json = a2g.iloc[num]
    info = ast.literal_eval(Json)['InformationList']['Information'][0]
    if 'GeneID' in info:
        gene_ids = info['GeneID']
        if (gene_ids != []):
            for each_id in gene_ids:
                row_to_add = pd.Series({'aid': a, 'GeneID': each_id})
                aid2gid = aid2gid.append(row_to_add, ignore_index=True)
aid2gid
```

```python
aid2gid.to_csv("assays_genes_associations.csv", index=False)
```
